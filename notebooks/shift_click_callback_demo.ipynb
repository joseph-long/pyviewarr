# Shift-click callback demo

This notebook demonstrates `ViewerConfig(on_shift_click=...)` and the optional `overlay_message`.
Shift-click inside the image to capture continuous (fractional) data-space coordinates.

```python
import numpy as np
import ipywidgets as widgets
import pyviewarr
```

```python
# Build a simple smooth image so subpixel coordinates are easy to see.
ny, nx = 300, 420
y, x = np.mgrid[0:ny, 0:nx]
img = (
    np.exp(-(((x - 130.0) / 30.0) ** 2 + ((y - 110.0) / 25.0) ** 2))
    + 0.6 * np.exp(-(((x - 300.0) / 40.0) ** 2 + ((y - 190.0) / 35.0) ** 2))
).astype(np.float32)
```

```python
marked_points = []

def on_shift_click(x: float, y: float) -> None:
    marked_points.append((x, y))
    with log_output:
        print(f"Shift-click at x={x:.3f}, y={y:.3f} (count={len(marked_points)})")

viewer_config = pyviewarr.ViewerConfig(
    on_shift_click=on_shift_click,
    overlay_message="Shift-click to mark points",
    cmap="inferno",
)

pyviewarr.show(img, width=600, height=300, viewer_config=viewer_config)
log_output = widgets.Output()
display(log_output)
```
