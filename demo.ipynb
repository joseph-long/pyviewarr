```python
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename

filename = get_pkg_data_filename('galactic_center/gc_msx_e.fits')
data = fits.getdata(filename)
```

```python
import pyviewarr
```

```python
viewer = pyviewarr.viewarr(data, stretch_mode='log', colormap='magma')
viewer
```

```python
config = viewer.get_viewer_config()
import dataclasses
dataclasses.asdict(config)
```

```python
viewer2 = pyviewarr.show(data, viewer_config=config)
viewer2
```
